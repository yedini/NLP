### BERT의 Masked Language Model 실습

```python
#!pip install transformers
```

```python
from transformers import TFBertForMaskedLM  # [MASK]로 되어있는 단어를 맞추기 위한 masked 언어 모델 구조로 BERT를 로드
from transformers import AutoTokenizer      # 해당 모델이 학습되었을 당시에 사용되었던 토크나이저 로드
```

```python
model = TFBertForMaskedLM.from_pretrained('bert-large-uncased')
tokenizer = AutoTokenizer.from_pretrained("bert-large-uncased")
```

#### 입력 예시

```python
# 예시 문장에 대해 정수 인코딩 수행
inputs = tokenizer('Soccer is a really fun [MASK].', return_tensors='tf')
```

```python
# 정수 인코딩 결과 확인
print(inputs['input_ids'])
```

```python
# 세그먼트 인코딩 결과 확인(문장 구분)
print(inputs['token_type_ids'])
```

```python
# 어텐션 마스크 확인: 패딩이 없으므로 1만 나온다
print(inputs['attention_mask'])
```

#### [MASK] 토큰 예측

```python
from transformers import FillMaskPipeline   # 모델과 토크나이저를 지정하면 masked 언어 모델의 예측 결과를 정리해서 보여주는 역할
pip = FillMaskPipeline(model=model, tokenizer=tokenizer)
```

```python
# [MASK] 위치에 들어갈 수 있는 상위 5개의 후보 단어 출력
pip('Soccer is a really fun [MASK].')
```

```python
pip('The Avengers is a really fun [MASK].')
```

```python
pip('I went to [MASK] this morning.')
```

### 한국어 BERT Masked Language Model 실습
klue/bert-base: 대표적인 한국어 BERT

```python
from transformers import TFBertForMaskedLM    # masked 언어 모델링을 위한 BERT 로드
from transformers import AutoTokenizer
```

```python
model = TFBertForMaskedLM.from_pretrained('klue/bert-base', from_pt=True)
tokenizer = AutoTokenizer.from_pretrained("klue/bert-base")
```

```python
# 예시 문장에 대해 정수 인코딩 수행
inputs = tokenizer('축구는 정말 재미있는 [MASK]다.', return_tensors='tf')
```

```python
# 세그먼트 인코딩 결과 확인
print(inputs['token_type_ids'])
```

```python
from transformers import FillMaskPipeline
pip = FillMaskPipeline(model=model, tokenizer=tokenizer)
```

```python
pip('축구는 정말 재미있는 [MASK]다.')
```

### BERT 다음 문장 예측
두 문장이 이어질 경우의 레이블은 0, 이어지지 않을 경우의 레이블은 1

```python
import tensorflow as tf
from transformers import TFBertForNextSentencePrediction  #두개의 문장이 이어지는 문장 관계인지 여부를 예측하기 위한 BERT
from transformers import AutoTokenizer
```

```python
model = TFBertForNextSentencePrediction.from_pretrained('bert-base-uncased')
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
```

#### 입력 예시

```python
# 실제로 이어지는 2개의 문장
prompt = "In Italy, pizza served in formal settings, such as at a restaurant, is presented unsliced."
next_sentence = "pizza is eaten with the use of a knife and fork. In casual settings, however, it is cut into wedges to be eaten while held in the hand."
```

```python
# 2개의 문장을 정수 인코딩
encoding = tokenizer(prompt, next_sentence, return_tensors='tf')
```

```python
# 정수 인코딩 결과
print(encoding['input_ids'])
```

이 때 101, 102는 특별토큰 CLS, SEP에 해당하는 토큰임

```python
print(tokenizer.cls_token, ':', tokenizer.cls_token_id)
print(tokenizer.sep_token, ':' , tokenizer.sep_token_id)
```

```python
# 인코딩한 결과를 다시 디코딩해보기
print(tokenizer.decode(encoding['input_ids'][0]))
```

```python
# 세그먼트 인코딩 결과 확인
print(encoding['token_type_ids'])
```

#### 다음 문장 예측하기

```python
logits = model(encoding['input_ids'], token_type_ids=encoding['token_type_ids'])[0] # 모델은 logit 값을 출력(softmax 함수 지나기 전의 값)
softmax = tf.keras.layers.Softmax()
probs = softmax(logits)  # 각 레이블에 대한 확률값
print(probs)
```

```python
# 더 큰 확률값을 가진 인덱스를 출력
print('최종 예측 레이블 :', tf.math.argmax(probs, axis=-1).numpy())
```

### 한국어 BERT 다음 문장 예측

```python
import tensorflow as tf
from transformers import TFBertForNextSentencePrediction
from transformers import AutoTokenizer
```

```python
# 한국어 bert 모델과 해당 모델을 학습할 당시에 사용한 tokenizer 불러오기
model = TFBertForNextSentencePrediction.from_pretrained('klue/bert-base', from_pt=True)
tokenizer = AutoTokenizer.from_pretrained("klue/bert-base")
```

```python
# 이어지는 두 개의 문장
prompt = "2002년 월드컵 축구대회는 일본과 공동으로 개최되었던 세계적인 큰 잔치입니다."
next_sentence = "여행을 가보니 한국의 2002년 월드컵 축구대회의 준비는 완벽했습니다."
encoding = tokenizer(prompt, next_sentence, return_tensors='tf')

logits = model(encoding['input_ids'], token_type_ids=encoding['token_type_ids'])[0]

softmax = tf.keras.layers.Softmax()
probs = softmax(logits)
print('최종 예측 레이블 :', tf.math.argmax(probs, axis=-1).numpy())
```

```python
print(tokenizer.cls_token, ':', tokenizer.cls_token_id)
print(tokenizer.sep_token, ':' , tokenizer.sep_token_id)
```

```python
# 상관없는 두 개의 문장
prompt = "2002년 월드컵 축구대회는 일본과 공동으로 개최되었던 세계적인 큰 잔치입니다."
next_sentence = "극장가서 로맨스 영화를 보고싶어요"
encoding = tokenizer(prompt, next_sentence, return_tensors='tf')

logits = model(encoding['input_ids'], token_type_ids=encoding['token_type_ids'])[0]

softmax = tf.keras.layers.Softmax()
probs = softmax(logits)
print('최종 예측 레이블 :', tf.math.argmax(probs, axis=-1).numpy())
```
