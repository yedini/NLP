## 1. 순환신경망(Recurrent Neural Network, RNN)
__입력과 출력을 시퀀스 단위로 처리하려는 모델.__      
- 은닉층 노드에서 활성화함수를 통해 나온 결과값을 출력층 방향으로도 보내서 다시 은닉층 노드 다음 계산의 입력으로 보낸다.

(하나의 네모는 벡터 단위)        
- 셀: RNN의 은닉층에서 활성화함수를 통해 결과를 내보내는 역할을 하는 노드. 메모리셀 또는 RNN 셀이라고 표현한다.     
바로 이전 시점의 은닉층 메모리셀에서 나온 값을 입력으로 사용한다(: 재귀적 활동!)
- hidden state(은닉상태): 메모리 셀이 출력층 방향으로, 또는 다음시점의 자신에게 보내는 값.

- 뉴런 단위의 RNN 시각화

=> 입력층의 뉴런 수 4, 은닉층 2, 출력층 2          
- RNN의 다양한 용도      
RNN은 입력과 출력 길이를 다르게 설계할 수 있어서 다양한 용도로 사용될 수 있다.

    - One-to-Many: Image Captioning 등에 사용(하나의 이미지 입력에 대해 사진의 제목을 출력)
    - Many-to-One: 감성 분류(긍정/부정), 스팸 메일 분류 등에 사용
    - Many-to-Many: 챗봇, 번역기, 개체명 인식, 품사 태깅 등에 사용
              
- RNN 수식

시점 t에서 hidden state를 $h_t$, input을 위한 가중치를 $W_x$, 이전 시점의 hidden state $h_(t-1) $을 위한 가중치를 $W_h$라고 할 때,         
은닉층: $h_t=tanh(W_xx_t+W_hh_(t-1) +b)$        
출력층: $y_t=f(W_yh_t+b)$        
- 배치 크기가 1일 때 은닉층의 연산
